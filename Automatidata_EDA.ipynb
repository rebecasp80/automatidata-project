```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,5)
```

```python
df = pd.read_csv('2017_Yellow_Taxi_Trip_Data.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
sns.histplot(df['fare_amount'], bins=50, kde=True)
plt.title('Distribución de tarifas')
plt.show()
```

```python
sns.histplot(df['fare_amount'], bins=50, kde=True)
plt.title('Distribución de tarifas')
plt.show()
```

```python
sns.boxplot(x=df['trip_distance'])
plt.title('Outliers en distancias')
plt.show()
```

```python
df['month'] = pd.to_datetime(df['tpep_pickup_datetime']).dt.month_name()
df['day'] = pd.to_datetime(df['tpep_pickup_datetime']).dt.day_name()
```

```python
rides_month = df['month'].value_counts().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])
sns.barplot(x=rides_month.index, y=rides_month.values)
plt.xticks(rotation=45)
plt.title('Número total de viajes por mes')
plt.show()
```

```python
rides_day = df['day'].value_counts().reindex([
    'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'
])
sns.barplot(x=rides_day.index, y=rides_day.values)
plt.title('Número total de viajes por día de la semana')
plt.show()
```

```python
revenue_day = df.groupby('day')['total_amount'].sum().reindex([
    'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'
])
sns.barplot(x=revenue_day.index, y=revenue_day.values)
plt.title('Ingresos totales por día de la semana')
plt.show()

revenue_month = df.groupby('month')['total_amount'].sum().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])
sns.barplot(x=revenue_month.index, y=revenue_month.values)
plt.xticks(rotation=45)
plt.title('Ingresos totales por mes')
plt.show()
```

```python
corr = df[['trip_distance','fare_amount','passenger_count','total_amount']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlaciones entre variables')
plt.show()
```

### Resultados y evaluación
He aprendido que las tarifas y distancias presentan outliers que deben limpiarse.
Los viajes son más frecuentes en días laborables y los ingresos aumentan los viernes.
El cliente probablemente querrá saber cómo afectan las horas punta y los recargos.
