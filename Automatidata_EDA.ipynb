```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,5)
```

```python
df = pd.read_csv('2017_Yellow_Taxi_Trip_Data.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
sns.histplot(df['fare_amount'], bins=50, kde=True)
plt.title('Distribución de tarifas')
plt.show()
```

```python
sns.histplot(df['fare_amount'], bins=50, kde=True)
plt.title('Distribución de tarifas')
plt.show()
```

```python
sns.boxplot(x=df['trip_distance'])
plt.title('Outliers en distancias')
plt.show()
```

```python
df['month'] = pd.to_datetime(df['tpep_pickup_datetime']).dt.month_name()
df['day'] = pd.to_datetime(df['tpep_pickup_datetime']).dt.day_name()
```

```python
rides_month = df['month'].value_counts().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])
sns.barplot(x=rides_month.index, y=rides_month.values)
plt.xticks(rotation=45)
plt.title('Número total de viajes por mes')
plt.show()
```

```python
rides_day = df['day'].value_counts().reindex([
    'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'
])
sns.barplot(x=rides_day.index, y=rides_day.values)
plt.title('Número total de viajes por día de la semana')
plt.show()
```

```python
revenue_day = df.groupby('day')['total_amount'].sum().reindex([
    'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'
])
sns.barplot(x=revenue_day.index, y=revenue_day.values)
plt.title('Ingresos totales por día de la semana')
plt.show()

revenue_month = df.groupby('month')['total_amount'].sum().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])
sns.barplot(x=revenue_month.index, y=revenue_month.values)
plt.xticks(rotation=45)
plt.title('Ingresos totales por mes')
plt.show()
```

```python
corr = df[['trip_distance','fare_amount','passenger_count','total_amount']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlaciones entre variables')
plt.show()
```

### Resultados y evaluación
He aprendido que las tarifas y distancias presentan outliers que deben limpiarse.
Los viajes son más frecuentes en días laborables y los ingresos aumentan los viernes.
El cliente probablemente querrá saber cómo afectan las horas punta y los recargos.

### Conclusión
EDA es importante porque permite detectar errores y patrones antes del modelado.
Las visualizaciones ayudan a comunicar hallazgos de forma clara y accesible.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv('/content/drive/MyDrive/Automatidata/2017_Yellow_Taxi_Trip_Data.csv')
```

```python
df.head()
```

```python
# Crear columna de duración de viaje
df['trip_duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']

# Revisar primeras filas
df[['tpep_pickup_datetime','tpep_dropoff_datetime','trip_duration']].head()
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
# Crear columna de duración de viaje
df['trip_duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']

# Revisar primeras filas
df[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_duration']].head()
```

```python
# Boxplot de distancia
sns.boxplot(x=df['trip_distance'], fliersize=1)
plt.title('Trip Distance - Outliers')

# Boxplot de importe total
sns.boxplot(x=df['total_amount'], fliersize=1)
plt.title('Total Amount - Outliers')
```

```python
# Histograma de distancias
sns.histplot(df['trip_distance'], bins=range(0,35,1))
plt.title('Distribución de distancias')

# Histograma de importes
sns.histplot(df['total_amount'], bins=range(-10,100,5))
plt.title('Distribución de importes')
```

```python
# Crear columnas de mes y día
df['month'] = df['tpep_pickup_datetime'].dt.month_name()
df['day'] = df['tpep_pickup_datetime'].dt.day_name()

# Conteo de viajes por mes
sns.barplot(x=df['month'].unique(), y=df['month'].value_counts())
plt.title('Número de viajes por mes')

# Conteo de viajes por día
sns.barplot(x=df['day'].unique(), y=df['day'].value_counts())
plt.title('Número de viajes por día')
```

```python
# Histograma de propinas por vendor
sns.histplot(data=df, x='tip_amount', hue='VendorID', multiple='stack')
plt.title('Distribución de propinas por vendor')

# Media de propinas por número de pasajeros
mean_tips = df.groupby('passenger_count')['tip_amount'].mean()
sns.barplot(x=mean_tips.index, y=mean_tips.values)
plt.axhline(df['tip_amount'].mean(), ls='--', color='red', label='Media global')
plt.legend()
plt.title('Propina media por número de pasajeros')
```

```python
# Media de distancia por drop-off
distance_by_dropoff = df.groupby('DOLocationID')['trip_distance'].mean().sort_values()
sns.barplot(x=distance_by_dropoff.index, y=distance_by_dropoff.values)
plt.title('Distancia media por drop-off location')
plt.xticks([])
```
